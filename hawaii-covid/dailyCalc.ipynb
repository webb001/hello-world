# Calculate the daily difference from the daily totals

reads totals.csv
write daily.csv

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("totals.csv",parse_dates=True)
```

```python
df["Date"] = pd.to_datetime(df["Date"],infer_datetime_format=True)
```

```python
df.set_index("Date",inplace=True)
```

```python
df.index
```

```python
df
```

```python
df1 = df.shift(1)
```

```python
df1
```

```python
dfDaily = df-df1
```

```python
dfDaily
```

```python
dfDaily.to_csv("daily.csv",float_format='%.0f',date_format='%m/%d/%Y')
```
