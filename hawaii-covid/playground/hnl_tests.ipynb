```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
df = pd.read_csv("hnl_tests.csv",parse_dates=True)
```

```python
df.set_index("Date",inplace=True)
```

```python
df = df.asfreq('D')
```

```python
df.describe()
```

```python
(df['Pos_s']/(df['Pos_s']+df['Neg_s'])*100).tail(7).describe()
```

```python
(df['Pos']/(df['Pos']+df['Neg'])*100).tail(7).describe()
```

```python
df['Pct_Pos'] = (df['Pos']/(df['Pos']+df['Neg'])*100)
```

```python
df['Pct_Pos_s'] = (df['Pos_s']/(df['Pos_s']+df['Neg_s'])*100)
```

```python
df[['Pct_Pos','Pct_Pos_s']].plot(figsize=(20,10),grid='on')
```

```python
(df['Pos']+df['Neg']).plot(kind='bar',figsize=(20,10))
```
