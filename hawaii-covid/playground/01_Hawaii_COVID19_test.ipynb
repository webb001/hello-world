# Short term / long term rate of change

```python
st = 7 # short term
lt = 14 # long term
```

```python
df.tail(lt).describe()
```

```python
df.tail(st).mean()/df.tail(lt).mean()
```

```python
R = df["Statewide total"].tail(st).mean()/df["Statewide total"].tail(lt).mean()
#R = df["Statewide total"].tail(st).quantile(q=0.5)/df["Statewide total"].tail(lt).quantile(q=0.5)
R
```

```python
D = df.index.values[-1]
D #type(D)
```

```python
#Starting = df.iloc[-1]["Statewide total"]
Starting = df["Statewide total"].tail(st).mean()
type(Starting)
```

```python
type(R*Starting)
```

```python
R*Starting
```

```python
D+np.timedelta64(1, "D")
```

```python
dates = pd.date_range(D+np.timedelta64(1, "D"), periods=st, freq='D')
```

```python
#values = [int(round(R*Starting))]
#for i in range(1,st):
#  values.append(int(round(values[i-1]*R)))
#values

values = [R*Starting]
for i in range(1,st):
  values.append(values[i-1]*R)
values

for i in range(len(values)):
  values[i] = int(round(values[i]))
values
```
